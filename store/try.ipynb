```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AAPL.csv')
```

```python
df[['year', 'month', 'day']] = df['date'].str.split(' ').str[0].str.split('-', expand=True)

# df['date'][0].split(' ')[0].split('-')
```

```python
data_2023 = df[df['year'] == '2023']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='open', data=data_2023, hue='month', palette='icefire')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='close', data=data_2023, hue='month', palette='icefire')
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='volume', data=data_2023, hue='month', palette='icefire')
```
